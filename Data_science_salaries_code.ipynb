```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import csv
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
#Read the csv file 
file_path = r"C:\Users\Kiruthicka\Desktop\ds_salaries.csv"
df = pd.read_csv(file_path, encoding='latin1')
```

```python
df
```

```python
# Specify the columns to be converted to integer type
columns_to_convert = ['Unnamed: 0', 'work_year', 'salary', 'salary_in_usd', 'remote_ratio']

# Print unique values in each column before conversion
for column in columns_to_convert:
    unique_values = df[column].unique()
    print(f"Unique values in {column}: {unique_values}")

# Convert specified columns to numeric type without removing missing values
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Fill only numeric values with 0 and convert to integer type
df[columns_to_convert] = df[columns_to_convert].fillna(0).astype(int)
```

```python
#Tail represents last 5 rows of the df 
df
```

```python
#Check for null values 
df.isnull().sum()
```

```python
#Remove null values
df.dropna(inplace=True)
```

```python
#Check for null values 
df.isnull().sum()
```

```python
#Replace 
df.replace({'employment_type': {'FT': 'Full Time', 'PT': 'Parttime'}}, inplace=True)
df.replace({'experience_level': {'MI': 'Middle level', 'SE': 'Senior level' , 'EN' :'Entry level'}}, inplace=True)
df.replace({'company_size': {'S': 'Small', 'M': 'Medium','L': 'Large'}}, inplace=True)
```

```python
#Drop columns
df.drop(['Unnamed: 0', 'remote_ratio', 'Hiring Manager', 'Work Schedule', 'Travel Requirements'], axis=1, inplace=True)
```

```python
#Salary Distribution:
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary (USD)')
plt.show()
```

```python
#Experience Level Distribution:
plt.figure(figsize=(10, 6))
sns.countplot(x='experience_level', data=df)
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

```python
#Employment Type Breakdown
plt.figure(figsize=(12, 8))
sns.violinplot(x='company_size', y='salary_in_usd', data=df)
plt.title('Violin Plot: Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
#Average Salary by Job Title
plt.figure(figsize=(12, 8))

# Bar plot
plt.subplot(1, 2, 1)
sns.barplot(x='salary_in_usd', y='job_title', data=df, ci=None)
plt.title('Average Salary by Job Title')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')

# Heatmap
plt.subplot(1, 2, 2)
heatmap_data = df.pivot_table(values='salary_in_usd', index='job_title', aggfunc='mean')
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".2f")
plt.title('Heatmap: Average Salary by Job Title')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')

plt.tight_layout()
plt.show()
```

```python
#Company Size Distribution
sizes = df['company_size'].value_counts()
labels = sizes.index

plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Company Sizes')
plt.show()
```
